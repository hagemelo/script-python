```python
from datetime import datetime
import pytz
import matplotlib.pyplot as plt
%matplotlib inline
import re
import pandas as pd
import numpy as np
import matplotlib.dates as dates
```

```python
def formatdatahora(x):
    return datetime.strptime(x[:-3], '%Y-%m-%d%H:%M')

def convertToint(val):
    if val:
        return float(val)/1000
    else:
        return 0
```

```python
COLUMNS = ['datetime', 'ip', 'method', 'uri', 'status', 'timetaken']
df_accesslog = pd.read_csv('u_ex190218.log',
                   sep='\s+',
                   engine='python', 
                   na_values='-', 
                   header=None ,        
                   usecols=[0,1,2,3, 10, 13],
                   names=COLUMNS ,
                   converters={'datetime': formatdatahora,
                              'timetaken': convertToint}
                   )
df_accesslog = df_accesslog.set_index('datetime')
#df_accesslog.head()
```

```python
df_SemEstaticos = df_accesslog.loc[((df_accesslog['uri'].str.contains('jpg'))==False) 
                              & ((df_accesslog['uri'].str.contains('png'))==False)
                              & ((df_accesslog['uri'].str.contains('js'))==False)
                              & ((df_accesslog['uri'].str.contains('css'))==False)]
#df_SemEstaticos.head()
```

```python
gp_hitLogin = df_SemEstaticos.loc[df_SemEstaticos['uri']=='/pcweb/login.aspx']
gp_hitLogin = gp_hitLogin.groupby(['datetime']).agg({    
    'uri': ['count']
})
gp_hitLogin =  gp_hitLogin.sort_values(by=[('uri', 'count')],ascending=False)
gp_hitLogin.head()
```
