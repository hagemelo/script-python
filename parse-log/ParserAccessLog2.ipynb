```python
from datetime import datetime
import pytz
import matplotlib.pyplot as plt
%matplotlib inline
import re
import pandas as pd

PATH = 'C:\\Users\\Inmetrics\\Documents\\vivo\\Help\\jvms\\JVMs\\Srv24\\'
FILE_NAME_EXCEL = 'Srv24_Parse.xlsx'

MENOR_05S = '0s - 0,5s'
ENTRE_05S_E_1S = '0,5s - 1s'
ENTRE_1S_E_15S = '1s - 1,5s'
ENTRE_15S_E_2S = '1.5s - 2s'
MAIOR_2S = 'Maior 2s'

def calc_hist(timetaken):
    if timetaken <= 0.5:
        return  MENOR_05S
    elif timetaken <= 1:
        return  ENTRE_05S_E_1S
    elif timetaken <= 1.5:
        return  ENTRE_1S_E_15S
    elif timetaken <= 2:
        return ENTRE_15S_E_2S
    else:
        return MAIOR_2S
```

```python
dadosAccesslog = pd.read_csv(PATH + 'access.log',
                   sep='\s',
                   engine='python', 
                   na_values='-', 
                   header=None,
                   names=['cip', 'date', 'time', 'csmethod', 'csuri' ,'status' ,'bytes', 'timetaken', 'Hist'],
                   )
calc_hist = [calc_hist(x) for x in dadosAccesslog['timetaken']]
dadosAccesslog['Hist'] = calc_hist
dadosAccesslog.head()
```

```python
grupURI = dadosAccesslog.groupby(by=['csuri'])['csmethod'].count().reset_index().sort_values(by='csuri')

df_filteredMENOR_05S = dadosAccesslog[(dadosAccesslog.Hist == MENOR_05S)]
df_filteredMENOR_05S = df_filteredMENOR_05S.groupby(by=['csuri'])['Hist'].count().reset_index().sort_values(by='csuri')

df_filteredENTRE_05S_E_1S = dadosAccesslog[(dadosAccesslog.Hist == ENTRE_05S_E_1S)]
df_filteredENTRE_05S_E_1S = df_filteredENTRE_05S_E_1S.groupby(by=['csuri'])['Hist'].count().reset_index().sort_values(by='csuri')

df_filteredENTRE_1S_E_15S = dadosAccesslog[(dadosAccesslog.Hist == ENTRE_1S_E_15S)]
df_filteredENTRE_1S_E_15S = df_filteredENTRE_1S_E_15S.groupby(by=['csuri'])['Hist'].count().reset_index().sort_values(by='csuri')

df_filteredENTRE_15S_E_2S = dadosAccesslog[(dadosAccesslog.Hist == ENTRE_15S_E_2S)]
df_filteredENTRE_15S_E_2S = df_filteredENTRE_15S_E_2S.groupby(by=['csuri'])['Hist'].count().reset_index().sort_values(by='csuri')

df_filteredMAIOR_2S = dadosAccesslog[(dadosAccesslog.Hist == MAIOR_2S)]
df_filteredMAIOR_2S = df_filteredMAIOR_2S.groupby(by=['csuri'])['Hist'].count().reset_index().sort_values(by='csuri')
```

```python
def contarHistMENOR_05S(p_pesq):
    if df_filteredMENOR_05S.loc[df_filteredMENOR_05S['csuri'] == p_pesq]['Hist'].values:
        result = df_filteredMENOR_05S.loc[df_filteredMENOR_05S['csuri'] == p_pesq]['Hist'].values[0]
    else:
        result = 0
    return result

def contarHistENTRE_05S_E_1S(p_pesq):
    if df_filteredENTRE_05S_E_1S.loc[df_filteredENTRE_05S_E_1S['csuri'] == p_pesq]['Hist'].values:
        result = df_filteredENTRE_05S_E_1S.loc[df_filteredENTRE_05S_E_1S['csuri'] == p_pesq]['Hist'].values[0]
    else:
        result = 0
    return result

def contarHistENTRE_1S_E_15S(p_pesq):
    if df_filteredENTRE_1S_E_15S.loc[df_filteredENTRE_1S_E_15S['csuri'] == p_pesq]['Hist'].values:
        result = df_filteredENTRE_1S_E_15S.loc[df_filteredENTRE_1S_E_15S['csuri'] == p_pesq]['Hist'].values[0]
    else:
        result = 0
    return result

def contarHistENTRE_15S_E_2S(p_pesq):
    if df_filteredENTRE_15S_E_2S.loc[df_filteredENTRE_15S_E_2S['csuri'] == p_pesq]['Hist'].values:
        result = df_filteredENTRE_15S_E_2S.loc[df_filteredENTRE_15S_E_2S['csuri'] == p_pesq]['Hist'].values[0]
    else:
        result = 0
    return result

def contarHistMAIOR_2S(p_pesq):
    if df_filteredMAIOR_2S.loc[df_filteredMAIOR_2S['csuri'] == p_pesq]['Hist'].values:
        result = df_filteredMAIOR_2S.loc[df_filteredMAIOR_2S['csuri'] == p_pesq]['Hist'].values[0]
    else:
        result = 0
    return result


```

```python
calc_MENOR_05S = [contarHistMENOR_05S(x) for x in grupURI['csuri']]

calc_ENTRE_05S_E_1S = [contarHistENTRE_05S_E_1S(x) for x in grupURI['csuri']]

calc_ENTRE_1S_E_15S = [contarHistENTRE_1S_E_15S(x) for x in grupURI['csuri']]

calc_ENTRE_15S_E_2S = [contarHistENTRE_15S_E_2S(x) for x in grupURI['csuri']]

calc_MAIOR_2S = [contarHistMAIOR_2S(x) for x in grupURI['csuri']]

grupURI[MENOR_05S] = calc_MENOR_05S
grupURI[ENTRE_05S_E_1S] = calc_ENTRE_05S_E_1S
grupURI[ENTRE_1S_E_15S] = calc_ENTRE_1S_E_15S
grupURI[ENTRE_15S_E_2S] = calc_ENTRE_15S_E_2S
grupURI[MAIOR_2S] = calc_MAIOR_2S
```

```python
grupURI.sort_values(by='csmethod', ascending=1)
grupURI.head()
```

```python
grupURI.columns = ['URI','QTD',MENOR_05S,ENTRE_05S_E_1S, ENTRE_1S_E_15S, ENTRE_15S_E_2S, MAIOR_2S]
grupURI.sort_values(by=MAIOR_2S, ascending=0).head()

writer = pd.ExcelWriter(PATH + FILE_NAME_EXCEL, engine='xlsxwriter')
grupURI.to_excel(writer, sheet_name='Plan1')
writer.save()
```
